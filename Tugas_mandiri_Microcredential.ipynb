#Latihan 5
Penjumlahan Matriks Ordo 7x9 nilainya didapatkan secara random


```python
matriksA = np.random.randint(1,5,(7,9))
matriksB = np.random.randint(4,10,(7,9))

print(matriksA)
print()
print(matriksB)

#Penjumlahan Matriks A dan B
for x in range(0, len(matriksA)):
  print()
  for y in range(0, len(matriksA[0])):
    print(matriksA[x][y] + matriksB[x][y], end='')
  print()
```
